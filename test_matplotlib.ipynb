```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 5, 0.1) #x = [0...5] in 0.1 chunks
y = np.sin(x) #y = sin(x)
plt.plot(x, y) #make a plot
plt.xlabel('x') #label x axis
plt.ylabel('sin(x)') #label y axis
plt.savefig('sinx.png', bbox_inches = "tight", dpi = 600)
```
